RunnablePassthrough is a core component in the LangChain Expression Language (LCEL) that passes its input directly through to the next step in a chain unchanged. It essentially acts as an identity function (f(x) = x), making it vital for preserving original user inputs, creating parallel execution paths, or dynamically restructuring data fields. Why It Is UsedIn a standard LangChain sequence, each component consumes the output of the previous step. If a step processes or transforms the data (e.g., a document retriever turning a search query into context text), the original user input can be completely lost before reaching the prompt template. RunnablePassthrough solves this problem by holding onto data streams without changing them.

## Pipeline RAG avançada

### 1. Reformulação da consulta para contextualizar

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import AIMessage, HumanMessage
from langchain_classic.chains import create_history_aware_retriever, create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

context_q_system_prompt = "Given the following chat history and the follow-up question which might reference context in the chat history, formulate a standalone question which can be understood without the chat history. Do NOT answer the question, just reformulate it if needed and otherwise return it as is."
context_q_user_prompt = "Question: {input}"
context_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", context_q_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", context_q_user_prompt)
    ]
)
```

Essa parte é relacionada a reescrita da pergunta original mas levando em consideração o histórico da conversa como contexto enriquecido. É importante notar que o prompt é uma solicitação de reescrita da pergunta original levando em conta o chat como histórico mas que o produto final seja uma solicitação em que você entenda o todo sem ter acesso ao histórico.

### 2. Chain para contextualização

```python
retriever
```

```python
history_aware_retriever = create_history_aware_retriever(
    llm = llm,
    retriever = retriever,
    prompt = context_q_prompt
)
```

#### create_history_aware_retriever

Esse retriever basicamente realiza a seguinte operação: toda vez que uma pergunta é feita vai ser recuperado todo o histórico das mensagens anteriores, a pergunta original vai ser reescrita com a LLM de tal forma que o histórico enriqueça o contexto da pergunta, e para que assim, sejam buscados documentos relacionados a pergunta e ao histórico.

##### Descrição

create_history_aware_retriever is a LangChain function that upgrades a basic RAG (Retrieval-Augmented Generation) system to handle conversational context. It acts as an LLM-powered query rewriter that looks at your previous conversation history to create standalone search terms.

##### Why You Need It

In a standard RAG pipeline, if a user asks a follow-up question like "What are its side effects?", the system won't know what "it" refers to.This retriever solves that problem by using a secondary language model to combine your current input with past conversation messages, generating a rewritten, fully-contextual search query (e.g., "What are the side effects of the medication mentioned previously?") before querying the vector database.

##### How It Works

1. With no chat history: The query is passed directly to your standard retriever.
2. With chat history: The function uses an LLM and a prompt to rewrite the user's latest question into a standalone query that understands the context of previous messages.
3. Retrieval: The newly rewritten query is sent to your vector store to fetch relevant documents.

### 3. Chain para perguntas e respostas (Q&A)

```python
qa_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "Pergunta: {input}\n\n Contexto: {context}")
    ]
)

qa_chain = create_stuff_documents_chain(llm, qa_prompt)

rag_chain = create_retrieval_chain(history_aware_retriever, qa_chain)
```

#### create_stuff_documents_chain

In LangChain, create_stuff_documents_chain is a helper function that combines a list of retrieved documents into a single string (often called "stuffing") and feeds them into a prompt for a Large Language Model (LLM). It is a fundamental component used in Retrieval-Augmented Generation (RAG) and summarization applications.

Cria uma *sub-chain* entre a LLM e o prompt principal de Q&A, já possuindo um espaço no prompt para o histórico de mensagens.

#### create_retrieval_chain

create_retrieval_chain is a LangChain helper function used to build Retrieval-Augmented Generation (RAG) applications. It automatically links a retriever (like a vector store) to a document-processing chain. It seamlessly passes user queries to fetch context, injects that context into a prompt, and hands it to a Large Language Model (LLM).

##### How It Works

1. Retrieve: The user's query is passed to the retriever, which pulls the most relevant documents from your data source.
2. Combine & Format: A document chain (often built using create_stuff_documents_chain) gathers the fetched documents and formats them into a single prompt.
3. Generate: The formatted prompt is sent to the LLM to generate the final, context-grounded response.

#### Conclusão

A parte do *stuffing* faz muito mais sentido quando você tem a função **create_stuff_documents_chain** sendo usada em conjunto com a função **create_retrieval_chain**. Na função **create_retrieval_chain** o *retriever* será usado para recuperar os documentos necessários levando em conta a pergunta original e histórico do chat e passará esses documentos para a sub-chain, desempenhada pela funcão **create_stuff_documents_chain**. Com os documentos em mãos será feito o stuffing gerando assim o prompt final para a resposta final.

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import AIMessage, HumanMessage
from langchain_classic.chains import create_history_aware_retriever, create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

def config_rag_chain(llm, retriever):
    # Prompt para contextualização
    context_q_system_prompt = "Given the following chat history and the follow-up question which might reference context in the chat history, formulate a standalone question which can be understood without the chat history. Do NOT answer the question, just reformulate it if needed and otherwise return it as is."
    context_q_user_prompt = "Question: {input}"
    context_q_prompt = ChatPromptTemplate.from_messages(
        [
            ("system", context_q_system_prompt),
            MessagesPlaceholder("chat_history"),
            ("human", context_q_user_prompt)
        ]
    )

    # Chain para contextualização
    history_aware_retriever = create_history_aware_retriever(
        llm = llm,
        retriever = retriever,
        prompt = context_q_prompt
    )

    # Prompt para Q&A
    system_prompt = """
        Você é um assistente virtual prestativo e está respondendo perguntas gerais sobre os serviços de uma empresa.
        Use os seguintes pedaços de contexto recuperado para responder à pergunta.
        Se você não sabe a resposta, apenas comente que não sabe dizer com certeza.
        Mas caso seja uma dúvida muito comum, pode sugerir como alternativa uma solução possível.
        Mantenha a resposta concisa.
        Responda em português. \n\n
    """

    qa_prompt = ChatPromptTemplate.from_messages([
        ("system", system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "Pergunta: {input}\n\n Contexto: {context}"),
    ])

    # Configurar LLM e Chain para perguntas e respostas (Q&A)

    qa_chain = create_stuff_documents_chain(llm, qa_prompt)

    rag_chain = create_retrieval_chain(
        history_aware_retriever,
        qa_chain,
    )

    return rag_chain
```

## Adicionando histórico das conversas

```python
store = {}
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
```

```python
qa_chain_with_history = RunnableWithMessageHistory(
    runnable=qa_chain,
    get_session_history=get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)
```

RunnableWithMessageHistory is a LangChain utility that automatically adds conversation memory to an AI chain. It wraps around your prompt and language model, tracking past messages and injecting them into the execution context so the AI remembers previous turns across multiple sessions.

**How it Works**

Instead of manually managing the history array, you pass it a function that fetches chat history based on a unique session_id.
1. Retrieval: Before the AI runs, it retrieves the conversation history for that specific session_id.
2. Appending: It appends the newest human prompt and the AI's response back into the history store.

## Configuração a função de Indexação e Recuperação

```python
from pathlib import Path
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

def config_retriever(folder_path="."):
    # Carregar documentos
    docs_path = Path(".")
    pdf_files = [f for f in docs_path.glob("*.pdf")]

    loaded_documents = [extract_text_pdf(pdf) for pdf in pdf_files]

    # Divisão em pedaçõs em texto / Split
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size = 1000,
        chunk_overlap = 200
    )
    chunks = []
    for doc in loaded_documents:
        chunks.extend(text_splitter.split_text(doc))
    
    # Embeddings
    embedding_model = "BAAI/bge-m3"

    embeddings = HuggingFaceEmbeddings(model_name=embedding_model)

    # Armazenamento
    vectorstore = FAISS.from_texts(chunks, embedding=embeddings)

    vectorstore.save_local('index_faiss')

    # Configurando o recuperador de texto / Retriever
    retriever = vectorstore.as_retriever(
        search_type = 'mmr',
        search_kwargs={'k':3, 'fetch_k':4}
    )

    return retriever
```

### Conclusão do projeto

```python
%%writefile app02.py
import streamlit as st
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from pathlib import Path
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import AIMessage, HumanMessage
from langchain_classic.chains import create_history_aware_retriever, create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_community.vectorstores import FAISS
from dotenv import load_dotenv
load_dotenv()

st.set_page_config(page_title="Atendimento SafeBank 🤖", page_icon="🤖")
st.title("Atendimento SafeBank")

id_model = "llama-3.3-70b-versatile"
temperature = 0.7
path = "."

### Carregamento da LLM
def load_llm(id_model, temperature):
  llm = ChatGroq(
    model=id_model,
    temperature=temperature,
    max_tokens=None,
    timeout=None,
    max_retries=2,
  )
  return llm

llm = load_llm(id_model, temperature)

### Exibição do resultado
def show_res(res):
  from IPython.display import Markdown
  if "</think>" in res:
    res = res.split("</think>")[-1].strip()
  else:
    res = res.strip()  # fallback se não houver tag
  display(Markdown(res))

### Extração do conteúdo
def extract_text_pdf(file_path):
  loader = PyMuPDFLoader(file_path)
  doc = loader.load()
  content = "\n".join([page.page_content for page in doc])
  return content

### Indexação e recuperação
def config_retriever():
  # Carregar documentos
  docs_path = Path(".")
  pdf_files = [f for f in docs_path.glob("*.pdf")]

  if len(pdf_files) < 1:
    st.error("Nenhum arquivo PDF carregado")
    st.stop()

  loaded_documents = [extract_text_pdf(pdf) for pdf in pdf_files]

  # Divisão em pedaços de texto / Split
  text_splitter = RecursiveCharacterTextSplitter(
      chunk_size=1000,
      chunk_overlap=200
  )
  chunks = []
  for doc in loaded_documents:
      chunks.extend(text_splitter.split_text(doc))

  # Embeddings
  embedding_model = "BAAI/bge-m3" #sentence-transformers/all-mpnet-base-v2

  embeddings = HuggingFaceEmbeddings(model_name=embedding_model)

  # Armazenamento
  vectorstore = FAISS.from_texts(chunks, embedding=embeddings)

  vectorstore.save_local('index_faiss')

  # Configurando o recuperador de texto / Retriever
  retriever = vectorstore.as_retriever(
      search_type='mmr',
      search_kwargs={'k':3, 'fetch_k':4}
  )

  return retriever

### Chain da RAG
def config_rag_chain(llm, retriever):

  # Prompt de contextualização
  context_q_system_prompt = "Given the following chat history and the follow-up question which might reference context in the chat history, formulate a standalone question which can be understood without the chat history. Do NOT answer the question, just reformulate it if needed and otherwise return it as is."

  context_q_system_prompt = context_q_system_prompt
  context_q_user_prompt = "Question: {input}"
  context_q_prompt = ChatPromptTemplate.from_messages(
      [
          ("system", context_q_system_prompt),
          MessagesPlaceholder("chat_history"),
          ("human", context_q_user_prompt),
      ]
  )

  # Chain para contextualização
  history_aware_retriever = create_history_aware_retriever(
    llm=llm, retriever=retriever, prompt=context_q_prompt
  )

  # Prompt para perguntas e respostas (Q&A)
  system_prompt = """Você é um assistente virtual prestativo e está respondendo perguntas gerais sobre os serviços de uma empresa.
  Use os seguintes pedaços de contexto recuperado para responder à pergunta.
  Se você não sabe a resposta, apenas comente que não sabe dizer com certeza.
  Mas caso seja uma dúvida muito comum, pode sugerir como alternativa uma solução possível.
  Mantenha a resposta concisa.
  Responda em português. \n\n"""

  qa_prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    MessagesPlaceholder("chat_history"),
    ("human", "Pergunta: {input}\n\n Contexto: {context}"),
  ])

  # Configurar LLM e Chain para perguntas e respostas (Q&A)

  qa_chain = create_stuff_documents_chain(llm, qa_prompt)

  rag_chain = create_retrieval_chain(
    history_aware_retriever,
    qa_chain,
  )

  return rag_chain

### Interação com chat
def chat_llm(rag_chain, input):

  st.session_state.chat_history.append(HumanMessage(content=input))

  response = rag_chain.invoke({
      "input": input,
      "chat_history": st.session_state.chat_history
  })

  res = response["answer"]
  res = res.split("</think>")[-1].strip() if "</think>" in res else res.strip()

  st.session_state.chat_history.append(AIMessage(content=res))

  return res

input = st.chat_input("Digite sua mensagem aqui...")

if "chat_history" not in st.session_state:
  st.session_state.chat_history = [
      AIMessage(content = "Olá, sou o seu assistente virtual! Como posso te ajudar?"),
  ]

if "retriever" not in st.session_state:
  st.session_state.retriever = None

for message in st.session_state.chat_history:
  if isinstance(message, AIMessage):
    with st.chat_message("AI"):
      st.write(message.content)
  elif isinstance(message, HumanMessage):
    with st.chat_message("Human"):
      st.write(message.content)

if input is not None:
  with st.chat_message("Human"):
    st.markdown(input)

  with st.chat_message("AI"):
    if st.session_state.retriever is None:
      st.session_state.retriever = config_retriever()
    rag_chain = config_rag_chain(llm, st.session_state.retriever)
    res = chat_llm(rag_chain, input)
    st.write(res)
```
