## Extração de conteúdos

```python
import os
import fitz # PyMuPDF

path = "Currículo1.pdf"

def extract_text_pdf(filepath):
    doc = fitz.open(filepath)
    text = ""
    for page in doc:
        text += page.get_text()
    doc.close()
    return text.strip()
```

```python
extract_text_pdf(path)
```

```python
extract_text_pdf("Currículo2.pdf")
```

## Carregando a LLM

```python
from langchain_groq import ChatGroq

def load_llm(id_model, temperature):
    llm = ChatGroq(
        model=id_model,
        temperature=temperature,
        max_tokens=None,
        timeout=None,
        max_retries=2
    )
    return llm
```

```python
id_model = "llama-3.3-70b-versatile"
temperature = 0.7

llm = load_llm(id_model, temperature)
```

## Testes iniciais

```python
prompt = f"""
Resuma esse currículo em 1 parágrafo:
{content}
"""
```

```python
prompt
```

```python
output = llm.invoke(prompt)
output.content
```

## Processamento e exibição do resultado

```python
def format_res(res, return_thinking=False):
    res = res.strip()

    if return_thinking:
        res = res.replace("<think>", "[pensando...]")
        res = res.replace("</think>", "\n---\n")
    else:
        if "</think>" in res:
            res = res.split("</think>")[-1].strip()
    
    return res

def show_res(res):
    from IPython.display import Markdown

    display(Markdown(res))
```

```python
res = format_res(output.content)

show_res(res)
```
